```python
rf1=RandomForestClassifier(random_state=8,
                                 min_samples_leaf=525,
                                 max_features=17,
                                n_estimators=23)
evm.eval_model(rf1,x_train,y_train,x_val,y_val)
```

```python
from sklearn.model_selection import cross_val_score
space = {
    'max_features': hp.choice('max_features',[3,4,5,17]),
    'min_samples_leaf': hp.quniform('min_samples_leaf', 500, 530, 1),
    'n_estimators':   hp.quniform('n_estimators', 20, 25, 1) 
}
space
```

```python
def objective(space):
   rf = RandomForestClassifier(max_features = space['max_features'],min_samples_leaf = int(space['min_samples_leaf']),n_estimators = int(space['n_estimators']) )
   acc=cross_val_score(rf, x_train,y_train,cv=50,scoring='roc_auc').mean()
   return{'loss': 1-acc, 'status': STATUS_OK }
best = fmin(
   fn=objective,  
   space=space,      
   algo=tpe.suggest,      
   max_evals=5
)
print("Best: ", best)
```

```python
best['max_features'],best['min_samples_leaf'],best['n_estimators']
```

```python
rf2 = RandomForestClassifier(max_features = best['max_features'],min_samples_leaf = int(best['min_samples_leaf']),n_estimators = int(best['n_estimators']) )
```

```python
evm.eval_model(rf2,x_train,y_train,x_val,y_val)
```
